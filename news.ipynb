```python
import requests

URL1 = 'https://edition.cnn.com/world/live-news/coronavirus-pandemic-vaccine-updates-03-11-21/index.html'
```

```python
URL2 = 'https://edition.cnn.com/world/live-news/coronavirus-pandemic-vaccine-updates-03-12-21/index.html'
```

```python
page1 = requests.get(URL1) 
```

```python
page2 = requests.get(URL2)
```

```python
print(page1.content)
```

```python
print(page2.content)
```

```python
from bs4 import BeautifulSoup

beautiful = BeautifulSoup(page1.content, 'html.parser')
soup = BeautifulSoup(page2.content, 'html.parser')
```

```python
news1_content = beautiful.find(id = 'posts-and-button')
news2_content = soup.find(id = 'posts-and-button')
```

```python

for link in news1_content.find_all('h2'):
    title = link.text.strip()
    print(f'Title: {title}')
```

```python
for link in news1_content.find_all('span'):
    date = link.text.strip()
    print(f'Date: {date}')
```

```python
for link in news1_content.find_all('p', {'class' : 'sc-dfVpRl fHIrNq'}):
    author = link.text.strip()
    print(f'Author: {author}')
```

```python
for link in news1_content.find_all('div', {'class' : 'sc-bdVaJa post-content-rendered render-stellar-contentstyles__Content-sc-9v7nwy-0 erzhuK'}):
    article = link.text.strip()
    print(f'Article: {article}')
    print('\r')
```

```python
news1_json = []

for link in news1_content.find_all('h2'):
    title = link.text.strip()
    news1_json.append({
        "title" : title
    })
    
for link in news1_content.find_all('span'):
    date = link.text.strip()
    news1_json.append({
        "date" : date
    })
    
for link in news1_content.find_all('p', {'class' : 'sc-dfVpRl fHIrNq'}):
    author = link.text.strip()
    news1_json.append({
        "author" : author
    })
    
for link in news1_content.find_all('div', {'class' : 'sc-bdVaJa post-content-rendered render-stellar-contentstyles__Content-sc-9v7nwy-0 erzhuK'}):
    article = link.text.strip()
    news1_json.append({
        "article" : article
    })
news1_json
```
